```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Exploratory Data Analysis

```python
# Gender values_count
gender_value_counts = df['Gender'].value_counts()
gender_value_counts
```

```python
plt.figure(figsize=(10,5))
explode = (0.03, 0.03)
gender_value_counts.plot.pie(autopct='%1.2f%%',explode = explode, fontsize=15)
plt.title('Gender Wise count')
plt.show()
```

```python
sns.countplot(data=df, x='Gender')
plt.title('Gender wise Value count')
plt.show()
```

```python
# plotting a bar chart for gender vs total amount
gender_from_total_amt = df.groupby(['Gender'])['Amount'].sum().reset_index()
gender_from_total_amt
```

```python
plt.figure(figsize=(7,4))
sns.barplot(x = gender_from_total_amt['Gender'], y=gender_from_total_amt['Amount'])
plt.title('Gender to total amount')
plt.show()
```

From the above graph most buyers are females and purchasing power of females greater than male.

```python
# age group
age_group_count = df['Age Group'].value_counts()
age_group_count
```

```python
ax = sns.countplot(data=df, x='Age Group', hue='Gender')
plt.figure(figsize=(10,4))

for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```

```python
# Total amount vs age group
amount_from_agegroup = df.groupby(['Age Group'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)
amount_from_agegroup
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=amount_from_agegroup, x='Age Group', y='Amount')
plt.show()
```

from above graph we can see that most of the buyers are age group between 26-35 yrs female.

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'])['Orders'].sum().reset_index().sort_values(by='Orders', ascending=False).head(10)
sales_state
```

```python
plt.figure(figsize=(16,5))
sns.barplot(x=sales_state['State'], y=sales_state['Orders'], color = 'pink')
plt.xlabel('State', fontsize=15)
plt.ylabel('Orders', fontsize=15)
plt.show()
```

from above graphs we can see most of the orders and total amount from uttar pradesh, Maharashtra, Karnataka

```python
# Marital Status
marital_status_count = df['Marital_Status'].value_counts()
marital_status_count
```

```python
ax = sns.countplot(data=df, x='Marital_Status', color='skyblue')
for bars in ax.containers:
  ax.bar_label(bars)
plt.show()
```

```python
# total amount from marital status
tot_amt_marital = df.groupby(['Marital_Status', 'Gender'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)
tot_amt_marital
```

```python
plt.figure(figsize=(7,4))
sns.barplot(data=tot_amt_marital, x='Marital_Status', y='Amount', hue='Gender')
plt.show()
```

from above graph we can see most of the buyers are married(women) and have high purchasing power
